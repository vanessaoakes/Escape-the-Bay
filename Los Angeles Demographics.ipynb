```python
import csv
import matplotlib.pyplot as plt
import pandas as pd

# Resource https://www.census.gov/quickfacts/fact/table/losangelescountycalifornia/PST045217
```

```python
la_df = pd.read_csv('Quick Facts - LA.csv')
la_df.drop(columns = ["Fact Note"], inplace=True)
la_df.drop(columns = ["Value Note for Los Angeles County, California"], inplace=True)

la_df.head(50)
```

```python
#Pulling Race Information
la_race = la_df.iloc[11:18,0:3]
la_race

```

```python
#Used Lambda to convert to Float
la_race['Los Angeles County, California'] = la_race['Los Angeles County, California'].apply(lambda x: float(x[:-1]))

#Removed ",percent"
la_race['Fact'] = la_race['Fact'].apply(lambda x: x[:-9])
la_race
```

```python
#Pie Chart + Pretty

#Want to change the label in a legend on the side
plt.pie(la_race['Los Angeles County, California'],labels=la_race['Fact'])
plt.show()
```
