```python
import csv
import matplotlib.pyplot as plt
import pandas as pd

# Resource: https://www.census.gov/quickfacts/fact/table/sanmateocountycalifornia/PST045217
```

```python
san_mateo_df = pd.read_csv('Quick Fact - San Mateo.csv')
san_mateo_df.drop(columns = ["Fact Note"], inplace=True)
san_mateo_df.drop(columns = ["Value Note for San Mateo County, California"], inplace=True)


san_mateo_df.head(100)
```

```python
#Pulling Race Information
san_mateo_race = san_mateo_df.iloc[11:18,0:3]
san_mateo_race


```

```python
#Used Lambda to convert to Float
san_mateo_race['San Mateo County, California'] = san_mateo_race['San Mateo County, California'].apply(lambda x: float(x[:-1]))

#Removed ",percent"
san_mateo_race['Fact'] = san_mateo_race['Fact'].apply(lambda x: x[:-9])
san_mateo_race
```

```python
#Pie Chart + Pretty

#Want to change the label in a legend on the side
plt.pie(san_mateo_race['San Mateo County, California'],labels=san_mateo_race['Fact'])
plt.show()


```

```python
#Age

san_mateo_age = san_mateo_df.iloc[7:10,0:3]
san_mateo_df.head()
```

```python
san_mateo_age['San Mateo County, California'] = san_mateo_age['San Mateo County, California'].apply(lambda x: float(x[:-1]))
san_mateo_age['Fact'] = san_mateo_age['Fact'].apply(lambda x: x[:-9])
san_mateo_age



```
