```python
import csv
import matplotlib.pyplot as plt
import pandas as pd

# Resource https://www.census.gov/quickfacts/fact/table/contracostacountycalifornia,sanmateocountycalifornia/PST045217
```

```python
contra_costa_df = pd.read_csv('Quick Facts - Contra Costa.csv')
contra_costa_df.drop(columns = ["Fact Note"], inplace=True)
contra_costa_df.drop(columns = ["Value Note for Contra Costa County, California"], inplace=True)

contra_costa_df.head(50)
```

```python
#Pulling Race Information
contra_costa_race = contra_costa_df.iloc[11:18,0:3]
contra_costa_race



```

```python
#Used Lambda to convert to Float
contra_costa_race['Contra Costa County, California'] = contra_costa_race['Contra Costa County, California'].apply(lambda x: float(x[:-1]))

#Removed ",percent"
contra_costa_race['Fact'] = contra_costa_race['Fact'].apply(lambda x: x[:-9])
contra_costa_race
```

```python
#Pie Chart + Pretty

#Want to change the label in a legend on the side
plt.pie(contra_costa_race['Contra Costa County, California'],labels=contra_costa_race['Fact'])
plt.show()



```

```python
#Age

contra_costa_age = contra_costa_df.iloc[7:10,0:3]
contra_costa_df.head()
```

```python
contra_costa_age['Contra Costa County, California'] = contra_costa_age['Contra Costa County, California'].apply(lambda x: float(x[:-1]))
contra_costa_age['Fact'] = contra_costa_age['Fact'].apply(lambda x: x[:-9])
contra_costa_age
```

```python
plt.pie(contra_costa_age['Contra Costa County, California'],labels=contra_costa_age['Fact'])
plt.show()
```
