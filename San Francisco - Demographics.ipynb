```python
import csv
import matplotlib.pyplot as plt
import pandas as pd

#Resource: https://www.census.gov/quickfacts/fact/table/sanfranciscocitycalifornia,US/PST045217#PST045217


```

```python
file_one_df = pd.read_csv('Quick Facts - SF.csv')

file_one_df.head(100)
```

```python
#Pulling Race Information
sf_race = file_one_df.iloc[11:18,0:3]
sf_race.drop(columns = ["Fact Note"], inplace=True)
sf_race
```

```python
#Used Lambda to convert to Float
sf_race['San Francisco city, California'] = sf_race['San Francisco city, California'].apply(lambda x: float(x[:-1]))

#Removed ",percent"
sf_race['Fact'] = sf_race['Fact'].apply(lambda x: x[:-9])
sf_race
```

```python
#Pie Chart + Pretty
#Want to change the label in a legend on the side
plt.pie(sf_race['San Francisco city, California'],labels=sf_race['Fact'])
plt.show()
```

```python
#Age

sf_age = file_one_df.iloc[7:10,0:3]
sf_age.drop(columns = ["Fact Note"], inplace=True)
sf_age
```
